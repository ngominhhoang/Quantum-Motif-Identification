```python
from subprocess import call
import math
import random

countt = 1
motif = "test"
#avg_degree = 5
for avg_degree in range(2, 11, 2):
    for num_nodes in range(200, 1200, 200):
        for t in range(0, 5, 1):
            #num_nodes = 20
            num_layers = 1 #fixed
            #(1 = Bifan, 2 = Biparallel, 3 = Cascade, 4 = FFL)
            if motif == "cascade":
                motif_type = 3
            elif motif == "bifan":
                motif_type = 1
            elif motif == "biparallel" or motif == "test":
                motif_type = 2
            elif motif =="FFL":
                motif_type = 4
            upper_bound = 0.35
            if avg_degree > 5:
                upper_bound = 0.15
            edge_ratio = round(random.uniform(0.05, upper_bound),2)
            planted_motifs = math.ceil(num_nodes*avg_degree*edge_ratio/(2*(4 - int(motif_type/3))))
            planted_motif_ratio = 1#fixed
            mutation_rate = 1#fixed
            graph_type = 2#1 = ER, 2 = BA
            path = 'dataset/synthetic/'+motif+'/'
            test_case = countt
            record = {'test_index': countt, 'nums_nodes': num_nodes, 'motif_type': motif, 'degree': avg_degree, 'edge_ratio':edge_ratio, 'planted_motifs': planted_motifs}
            list_records.append(record)
            countt = countt + 1
            command = ['./graph_generator/SyntheticGraphGenerator-multipleTest',
                       str(num_nodes), str(num_layers), str(motif_type), str(avg_degree),
                       str(edge_ratio), str(planted_motif_ratio), str(mutation_rate), str(graph_type), str(path), str(test_case)]
            call(command)


        #break
```

```python
import os
import copy

saved_path = 'dataset/synthetic/'+motif+'/list_records.pkl'
with open(saved_path, "rb") as fp:   # Unpickling
    list_records = pickle.load(fp)

ratio_list = [0.2, 0.5, 0.8]
countt = 1
new_list_records = []
weighted_path = 'dataset/synthetic/'+motif+'_weighted/'

for record in list_records:
    input_path = 'dataset/synthetic/'+motif+'/'+str(record['test_index'])+'.txt'
    
    for ratio in ratio_list:
        #print(countt)
        new_record = copy.copy(record)
        new_record['test_index'] = countt
        countt = countt + 1
        new_record['activation_ratio'] = ratio
        new_list_records.append(new_record)
        output_path = os.path.join(weighted_path, str(new_record['test_index'])+'.txt')
        input_file = open(input_path, "r")
        output_file = open(output_path, "w")
        print(input_path,' ', output_path)
        edge_list = []
        while True:
            s = input_file.readline()
            if s == '':
                break
            ed = s[:-1].split(' ')
            # w = 0 => activation, w = 1 => inhibition
            if random.uniform(0, 1) < ratio:
                w = 0
            else:
                w = 1
            output_file.write(ed[0]+' '+ed[1]+' '+str(w)+'\n')
            #break
        #break
    #break
```
